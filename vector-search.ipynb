```python
books["tagged_description"].to_csv("tagged_description.txt",
                                   sep = "\n",
                                   index = False,
                                   header = False)
```

```python
raw_documents = TextLoader("tagged_description.txt").load()
text_splitter = CharacterTextSplitter(chunk_size=1, chunk_overlap=0, separator="\n")
documents = text_splitter.split_documents(raw_documents)
```

```python
documents[0]
```

```python
import os
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI

# 1. API Anahtarını yükle
load_dotenv()

print("API Anahtarı .env dosyasından yüklendi.")
print("Google sohbet modeli (gemini-1.5-flash) test ediliyor...")

try:
    # Sohbet modelini başlat (YENİ MODEL ADI!)
    llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")

    print("Model başarıyla yüklendi.")

    # Basit bir test isteği
    response = llm.invoke("Merhaba, adın ne?")

    print("\n--- TEST BAŞARILI ---")
    print("Google'dan yanıt geldi:")
    print(response.content)

except Exception as e:
    print("\n--- TEST BAŞARISIZ ---")
    print("Sohbet modeli çalışmadı. Alınan Hata:")
    print(e)
```

```python

from langchain_google_genai import GoogleGenerativeAIEmbeddings
# Google'ın embedding modelini başlat

# ("models/embedding-001" ücretsiz modelin adıdır)
google_embeddings = GoogleGenerativeAIEmbeddings(model="models/text-embedding-004")



# Chroma veritabanını bu model ile oluştur

db_books = Chroma.from_documents(

documents,

embedding=google_embeddings # <-- 3. Değiştirilmiş model

)
```

```python
query = "A book to teach children about nature"
docs = db_books.similarity_search(query, k = 10)
docs
```

```python
books[books["isbn13"] == int(docs[0].page_content.split()[0].strip())]
```

```python
def retrieve_semantic_recommendations(
        query: str,
        top_k: int = 10,
) -> pd.DataFrame:
    recs = db_books.similarity_search(query, k = 50)

    books_list = []

    for i in range(0, len(recs)):
        books_list += [int(recs[i].page_content.strip('"').split()[0])]

    return books[books["isbn13"].isin(books_list)]
```

```python
retrieve_semantic_recommendations("A book to teach children about nature")
```
